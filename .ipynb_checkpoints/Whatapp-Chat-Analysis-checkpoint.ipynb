```python
import re
import pandas as pd 
```

```python
f=open("WhatsApp Chat with CSE-B, IIIT'26 unofficial.txt",'r',encoding='utf-8')
```

```python
data=f.read()
```

```python
print(data)
```

```python
pattern='\d{1,2}\/\d{1,2}\/\d{2},\s\d{1,2}:\d{2}\s[AP]M\s-\s'
```

```python
messages=re.split(pattern,data)[1:]
messages
```

```python
dates=re.findall(pattern,data)
dates
```

```python
for i in range(len(dates)):
    dates[i]=dates[i].replace('\u202f',' ')
dates
```

```python
df=pd.DataFrame({'user_message':messages,'message_date':dates})
df['message_date'] = df['message_date'].str.rstrip(' -')
df['message_date'] = pd.to_datetime(df['message_date'], format='%m/%d/%y, %I:%M %p')
df.rename(columns={'message_date':'date'},inplace=True)
df.sample(5)
```

```python
df.shape
```

```python
    # Creating new user and message columns
    users = []
    messages = []
    for message in df['user_message']:
        entry = re.split('([\w\W]+?):\s', message)
        if entry[1:]:  # user name
            users.append(entry[1])
            messages.append(" ".join(entry[2:]))
        else:
            users.append('group_notification')
            messages.append(entry[0])

    df['user'] = users
    df['message'] = messages
    df.drop(columns=['user_message'], inplace=True)
```

```python
f=open('stop_hinglish.txt','r')
stop_word=f.read()
print(stop_word)
```

```python
words=[]

for message in temp['message']:
    for word in message.lower().split(' '):
        if word not in stop_word:
            words.append(word)
```
